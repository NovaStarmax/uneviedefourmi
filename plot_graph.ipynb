```python
import matplotlib.pyplot as plt
import networkx as nx
from txt_to_graph import *

# Registre pour conserver chaque graphe et sa position
GRAPHS = {}
```

Display options

```python
options = dict(
    with_labels=True,
    node_size=1500,
    node_color="lightblue",
    arrowsize=20,
    font_size=12
)
```

# F0

```python
G = nx.DiGraph([
    ("Sv", "S1"),  
    ("Sv", "S2"),  
    ("S1", "Sd"),  
    ("S2", "Sd"),
])

pos = {
    "Sv": (0, 0), 
    "S1": (1, 1),  
    "S2": (1, -1),  
    "Sd": (2, 0), 
}

# Sauvegarde
GRAPHS["F0"] = (G, pos)

nx.draw_networkx(G, pos, **options)

plt.axis("off")
plt.show()
```

```python
labels = ["F0", "F1", "F2", "F3", "F4", "F5"]
fig, axes = plt.subplots(3, 2, figsize=(14, 18))
axes = axes.ravel()
for ax, key in zip(axes, labels):
    G, pos = GRAPHS.get(key, (None, None))
    if G is None:
        ax.set_title(f"{key} (non défini)")
        ax.axis('off')
        continue

    colors = [("#34a853" if n=='Sv' else ("#ea4335" if n=='Sd' else "#80c0ff")) for n in G.nodes]
    nx.draw_networkx(G, pos=pos, ax=ax, with_labels=True, node_size=1200, node_color=colors, arrowsize=18, font_size=11)
    ax.set_title(key)
    ax.axis('off')

plt.tight_layout()
plt.show()
```
